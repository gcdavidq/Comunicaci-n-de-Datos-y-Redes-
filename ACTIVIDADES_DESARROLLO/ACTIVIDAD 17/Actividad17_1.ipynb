**Problema 1: Simulación de un cliente-servidor**

Conceptos: Cliente, Servidor, Protocolo, TCP/IP

Escribe una función en Python que simule la comunicación entre un cliente y un servidor usando la programación funcional. El cliente debería ser capaz de enviar mensajes y el servidor de responder de acuerdo con un protocolo simple (por ejemplo, eco, reverso del mensaje, etc.).

Implementa funciones separadas para el comportamiento del cliente y del servidor.
Simula el envío y recepción de mensajes.
Aplica conceptos de TCP/IP en un contexto abstracto.

*Paso 1: Definir las funciones del servidor:*

```python
def servidor(mensaje):
    """Procesa los mensajes enviados por el cliente según el comando especificado."""
    comando, _, contenido = mensaje.partition(':')
    if comando.strip().lower() == 'eco':
        return contenido.strip()
    elif comando.strip().lower() == 'reverso':
        return contenido.strip()[::-1]
    else:
        return "Comando no reconocido."
```

*El servido debe ser capaz de interpretar el mensaje recibido y responder de acuerdo a un comando especificado en el mensaje.*

*Paso 2: Definir el cliente texto en cursiva*

```python
def cliente(mensaje):
    """Envía mensajes al servidor y procesa la respuesta."""
    respuesta = servidor(mensaje)
    print(f"Servidor dice: {respuesta}")
```

*Paso 3: Simulación de la comunicación Para simular la comunicación, simplemente llamaremos a la función cliente con diferentes mensajes:*

```python
def main():
    cliente("eco: Hola Mundo")
    cliente("reverso: Hola Mundo")
    cliente("sumar: 123 + 456")  # Este debería dar un error ya que no está implementado

if __name__ == "__main__":
    main()
```

Simular retrasos de Red: Introduce retrasos aleatorios para simular la latencia de la red y observa cómo afecta la comunicación.

```python
import random
import time
def servidor(mensaje): #definimos la funcion servidor y agregamos como parametro mensaje
  """Procesa los mensajes enviados por el cliente segun el comando especificado."""
  request=input("Estás seguro de querer procesar el mensaje y convertirlo en diferentes formatos:")
  if request=="si" or request=="SI" or request=="Sí" or request=="Si": #creamos una condicional general que nos dirá si quieremos procesar el mensaje en el servidor o no
    comando,_,contenido=mensaje.partition(':') #dividimos el mensaje en 3 partes usando el caracter :
    if comando.strip().lower()=='eco': #verificamos si existe la palabra eco, que devolverá el mensaje requerido
      return retraso(contenido.strip()) #agregamos la funcion de regraso, que toma como parametro el contenido del mensaje.
    elif comando.strip().lower()=='reverso': #verificamos si existe la palabra reverso como comando
      return retraso(contenido.strip()[::-1]) #en caso exista, llamamos a la funcion retraso y tomamos como parametro el contenido del mensaje al reves
    elif comando.strip().lower()=='sumar':
      return retraso(suma(contenido)) ##llamar a la funcion que realiza la suma con su retraso
    elif comando.strip().lower()=='may':
      return retraso(contenido.strip().upper()) #convertir el mensaje a mayuscula, con retraso
    else:
      return "Comando no reconocido"
  else:
    print("No se realizó ninguno proceso")
```

```python
def suma(contenido): #crear la variable suma y agregar el parametro contenido
  valor1,_,valor2=contenido.partition('+') #realizar una division del mensaje empleando como separador el signo +
  valor1=int(valor1.strip()) #convertimos el valor de la variable valor1 a entero y eliminamos los espacios en blanco
  valor2=int(valor2.strip()) #convertimos el valor de la variable valor2 a entero y eliminamos los espacios en blanco
  return valor1 + valor2 #retornamos la suma de ambas variables
```

```python
def retraso(respuesta): #Generar un retraso aleatorio
  demora=random.uniform(5,10) #generamos numeros random entre 5 y 10
  time.sleep(demora) #convertimos a variable de tiempo los numeros random generados
  return respuesta
```

```python
def cliente(mensaje):
    """Envía mensajes al servidor y procesa la respuesta."""
    respuesta = servidor(mensaje)
    print(f"Servidor dice: {respuesta}")
```

```python
def main():
    cliente("eco: Hola Mundo")
    cliente("reverso: Hola Mundo")
    cliente("sumar: 123 + 456")
    cliente("may: hola mundo")

if __name__ == "__main__":
    main()
```
